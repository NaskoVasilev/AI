```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report
```

# Logistic regression
### Classification

```python
diabetes_data = pd.read_csv("diabetes_data/diabetic_data.csv")
```

```python
diabetes_data
```

```python
diabetes_data.shape
```

```python
diabetes_data.dtypes
```

```python
diabetes_target = diabetes_data["readmitted"]
diabetes_attributes = diabetes_data.drop("readmitted", axis = 1)
```

```python
diabetes_attributes = pd.get_dummies(diabetes_attributes)
diabetes_attributes.shape
```

```python
diabetes_attributes_scaled = MinMaxScaler().fit_transform(diabetes_attributes)
```

```python
logistic_regression_base = LogisticRegression(C = 1e9)
logistic_regression_base.fit(diabetes_attributes_scaled, diabetes_target)
logistic_regression_base.score(diabetes_attributes_scaled, diabetes_target)
```

```python
logistic_regression_normalized = LogisticRegression(C = 1)
logistic_regression_normalized.fit(diabetes_attributes_scaled, diabetes_target)
logistic_regression_normalized.score(diabetes_attributes_scaled, diabetes_target)
```

```python
logistic_regression_high_normalized = LogisticRegression(C = 0.001)
logistic_regression_high_normalized.fit(diabetes_attributes_scaled, diabetes_target)
logistic_regression_high_normalized.score(diabetes_attributes_scaled, diabetes_target)
```

```python
diabetes_attributes_train, diabetes_attributes_test, diabetes_target_train, diabetes_target_test = train_test_split(diabetes_attributes, diabetes_target, test_size=10000, random_state=30, stratify=diabetes_target)
```

```python
diabetes_target_test.groupby(diabetes_target_test).size() / len(diabetes_target_test)
```

```python
diabetes_target_train.groupby(diabetes_target_train).size() / len(diabetes_target_train)
```

```python
logistic_regression_split = LogisticRegression()
logistic_regression_split.fit(diabetes_attributes_train, diabetes_target_train)
```

```python
logistic_regression_split.score(diabetes_attributes_test, diabetes_target_test)
```

```python
diabities_predictions_test = logistic_regression_split.predict(diabetes_attributes_test)
```

```python
print(classification_report(diabetes_target_test, diabities_predictions_test))
```

```python
params = {
    "C": [0.01, 0.1, 1, 10, 100, 1000],
    "fit_intercept": [True, False],
    "max_iter": [10, 100, 1000]
}
gird_search = GridSearchCV(estimator=LogisticRegression(), param_grid = params)
```

```python
gird_search.fit(diabetes_attributes_train[:100], diabetes_target_train[:100])
```

```python
gird_search.best_estimator_
```

```python
gird_search.best_estimator_.score(diabetes_attributes_test, diabetes_target_test)
```

```python
print(classification_report(diabetes_target_test, gird_search.best_estimator_.predict(diabetes_attributes_test)))
```
